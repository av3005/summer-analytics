### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('cars.csv')

avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")


sns.lineplot(
    data=avg_mpg_by_year_origin,
    x='model_year',
    y='mpg',
    hue='origin',
    marker='o',
    palette='Set2'
)

plt.title('Evolution of Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.xticks(sorted(df['model_year'].unique()))  
plt.tight_layout()

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('cars.csv')


plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    palette='Set2',
    hue_order=['japan', 'europe', 'usa'],
    edgecolor='gray',
    alpha=0.7
)

plt.title('Horsepower vs Weight\nColor by Origin, Size by MPG', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
car_year_counts = df.groupby('name')['model_year'].nunique()
repeated_models = car_year_counts[car_year_counts > 1].index
filtered = df[df['name'].isin(repeated_models)]
mpg_std = filtered.groupby('name')['mpg'].std()
consistent_models = mpg_std[mpg_std < 1.0].index
final = filtered[filtered['name'].isin(consistent_models)]
result = final.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    avg_mpg=('mpg', 'mean')
).reset_index()
result = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])
print(result)
```
